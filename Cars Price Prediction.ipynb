```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
cars = pd.read_csv('car data.csv')
```

```python
cars.head()
cars.shape
```

# Preprocessing

```python
cars.drop(columns = ['Owner', 'Year'], axis = 1, inplace = True)
```

```python
cars['Fuel_Type'].value_counts()
```

```python
cars = cars[cars['Fuel_Type'] != 'CNG']
```

```python
cars['Transmission'].replace({'Manual': 0, 'Automatic': 1},inplace = True)
cars['Fuel_Type'].replace({'Petrol': 0, 'Diesel': 1},inplace = True)
cars['Seller_Type'].replace({'Individual': 0, 'Dealer': 1},inplace = True)
```

```python
cars.head()
```

```python
cars.Car_Name = cars.Car_Name.astype('category').cat.codes
```

```python
cars.head()
```

```python
cars.shape
```

# Splitting The Data

```python
X = cars.drop(columns = ['Selling_Price'])
y = cars['Selling_Price']

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```
